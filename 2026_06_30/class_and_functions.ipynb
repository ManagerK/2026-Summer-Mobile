## 코틀린의 여러가지 클래스 활용 및 람다 함수 학습

```kotlin
open class Shape {
    open fun draw() {
        println("도형 그리기")
    }
}

open class Ellipse : Shape() {
    override fun draw() = println("타원형 그리기")
}

open class Circle : Shape() {
    override fun draw() {
        println("Shape을 상속 받아 : 원 그리기")
    }
}

class Rect : Shape() {
    override fun draw() = println("사각형 그리기") 
    
    fun info() {
        println("저는 사각형입니다.")
    }
}

class RoundCircle : Circle() 

fun main() {
    val shapes: List<Shape> = listOf(Circle(), Rect(), RoundCircle(), Circle(), Rect(), Circle(), Ellipse())
    
    shapes.forEach { it.draw() }
    
    println("사각형 검사 : ${shapes[1] is Rect}")
    if (shapes[1] is Rect) {
        val r = shapes[1] as Rect
        r.info()
        println(r::class)
    }
    
    for (s in shapes) {
        when (s) {
            is Circle -> println("원 객체")
            is Rect -> println("사각형 객체")
            is Ellipse -> println("타원형 객체")
            is Shape -> println("** 도형 객체 **")
            else -> {
                println("내가 모르는 객체")
            }
        }
    }
}

main()
```

### 추상 클래스


```kotlin
abstract class Animal1

abstract class Animal2 {
    abstract fun bark()
    fun run(name: String) = println("${name}이 달립니다.")
}

class Dog : Animal1()

class Cat : Animal2() {
    override fun bark() = println("야옹...")
    fun run() = println("야옹이가 달리는데//")
}

fun main() {
    val dog = Dog()
    
    val cat = Cat()
    cat.bark()
    cat.run()
    cat.run("메롱")
}

main()
```

## interface

```kotlin
interface Flyable {
    fun fly()
    fun land() = println("착륙")
}

class Duck : Flyable {
    override fun fly() = println("오리가 날다")
    override fun land () = println("오리가 앉다")
}
```

```kotlin

    val duck = Duck()
    duck.fly()
    duck.land()
```

## 클래스 위임

```kotlin
interface Printer {
    fun print(msg: String)
}

class ColorPrinter : Printer {
    override fun print(msg: String) = println("[color] $msg")
}

class Logger1 : Printer {
    override fun print(msg: String) = println("[BW] $msg")
    
    fun log(msg: String) {
        this.print(msg)
    }
}

fun main() {
    val logger1 = Logger1()
    logger1.log("시작")
}

main()
```

```kotlin
val color_printer = ColorPrinter()
color_printer.print("시작")
```

```kotlin
class Logger2(private val printer: Printer ) : Printer {
    override fun print(msg: String) = println("[BW] $msg")
    fun log(msg: String) {
        print(msg)
    }
}

val logger2 = Logger2(color_printer)
logger2.log("시작")
```

### 람다 함수

```kotlin
val double : (Int) -> Int = { it * 2 }
val d = double(4)
println(d)
```

```kotlin
fun multiplier(factor: Int): (Int) -> Int = { it * factor }

val unknown = multiplier(3)
println("함수 동작 : ${unknown(5)}")
```

```kotlin
val some = {no : Int -> println("$no 값 입력")}
some(10)
```

```kotlin
val some2 : (Int) -> Unit = { println("$it 입력")}
some2(10)
```

```kotlin
fun testFun(no: Int, arg: (Int)->Boolean ): ()->String {
    val result = if(arg(no)) { // (Int)->Boolean, true/false
        "valid"
    } else {
        "invalid"
    }
    return { "testFun result: $result" } // ( () -> "..." ) => ( -> "..." ) => { "..." }
}

val testResult = testFun(-10, {no -> no>0})
testResult()
```
