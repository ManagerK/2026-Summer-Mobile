### 함수의 기본값 정의
- 함수 인수에 값을 지정하면, 가변 인수를 전달 할 수 있음

```kotlin
import kotlin.math.pow
fun power(base:Double, exp:Int =2): Double {
    return base.pow(exp)
}

println("2^3 = ${power(2.0, 3)}")
println("3^2 = ${power(3.0)}")
```

## when 제어문 사용
- when 제어문은 기존의 switch 문보다 강력함
- 일반적인 switch 문은 if 동등절에 해당하나 when제어문은 범위 및 다양한 조건을 포함할 수 있음

```kotlin
//println("입력스트링: ")
//val input: String = readln()

val data: Any = "Test"

when(data) {

    20, 30 -> println("입력 값이 20 또는 30임")
    in 1..10 -> println("입력 값이 1이상 10이하") 
    is String -> println("입력이 String임")
    else -> println("입력 값이 예상을 벗어남")
}
```

## 교수님 퀴즈 문제 풀어보기

```kotlin
 
    // 교수님이 제시한 100, 1000, 5000번 실행 및 x 출력
    val x1 = series(100)
    val x2 = series(1000)
    val x3 = series(5000)

    println("series(100)  -> x = $x1")
    println("series(1000) -> x = $x2")
    println("series(5000) -> x = $x3")


fun series(n: Int): Double {
    var sum = 0.0
    
    for (i in 0 until n) {
        
        val denominator = 2.0 * i + 1.0
        
        if (i % 2 == 0) {
            sum += 1.0 / denominator
        } else {
            sum -= 1.0 / denominator
        }
    }
    
    return sum * 4
}
```

## 반복문을 통해서 원주율 계산하기
- p1/4 = 1 - 1/3 + 1/5 - 1/7 + ...

```kotlin
fun series(n: Int) : Double {
    var x : Double = 0.0
    for(i in 1..n) {
        val denom = 2 * i - 1
        val sign = if (i % 2 == 1) 1.0 else -1.0
        x += sign / denom.toDouble()
    }
    return x * 4
}

println("series(100) = ${series(100)}")
println("series(1000) = ${series(1000)}")
println("series(5000) = ${series(5000)}")
```
